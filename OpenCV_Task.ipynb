```python
import cv2
import numpy as np

def rotate_image(image, angle):
    height, width = image.shape[:2]
    rotation_matrix = cv2.getRotationMatrix2D((width / 2, height / 2), angle, 1)
    rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))
    grayscale = cv2.cvtColor(rotated_image,cv2.COLOR_BGR2GRAY)
    return grayscale

def save_rotated_images(input_image_path, output_folder, num_angles):
    original_image = cv2.imread(input_image_path)

    # Create output folder if it doesn't exist
    import os
    os.makedirs(output_folder, exist_ok=True)

    for angle in range(0, 360, int(360/num_angles)):
        rotated_image = rotate_image(original_image, angle)
        output_path = f"{output_folder}/rotated_{angle}.jpg"
        cv2.imwrite(output_path, rotated_image)
        print(f"Saved {output_path}")

# Specify input image and output folder
input_image_path = "ghayur.jpg"
output_folder = "output_folder"

# Specify the number of angles to cover the full 360 degrees
num_angles = 360  # You can adjust this based on the desired granularity

# Generate and save rotated images
save_rotated_images(input_image_path, output_folder, num_angles)
```
