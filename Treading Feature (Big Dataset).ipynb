# Plotting Data

```python
plt.figure(figsize=(9,8))
plt.plot(df.index,df['last_price'])
plt.title('Stock Price')
plt.xlabel('Time')
plt.ylabel('Last_Price')
plt.show()
```

```python
rolling_mean = df['last_price'].rolling(window=5).mean()
rolling_std= df['last_price'].rolling(window=5).std()
rolling_sum= df['last_price'].rolling(window = 5).sum()
```

```python
rolling_mean
```

```python
rolling_sum
```

```python
rolling_std
```

```python
plt.plot(df.index,df['last_price'],label = 'original data')
plt.plot(df.index, rolling_mean, label = 'Rolling mean data',color='r')
plt.legend()
plt.show()
```

```python
plt.plot(df.index,df['last_price'],label = 'original data')
plt.plot(df.index, rolling_std, label = 'Rolling std data',color='r')
plt.legend()
plt.show()
```

```python
plt.plot(df.index,df['last_price'],label = 'original data')
plt.plot(df.index, rolling_sum, label = 'Rolling Sum data',color='r')
plt.legend()
plt.show()
```

```python
df['Label'] = df['last_price'].diff()
df['Label'] = (df['Label']>0).astype(int)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,mean_absolute_error,mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
df = df.dropna()
```

```python
x = df.drop(['last_price','Label','date_format','id'],axis = 'columns')
y = df.Label
```

```python
#standard_scaler = StandardScaler()
#x = standard_scaler.fit_transform(x)
```

```python
y
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size = 0.3)
```

```python
Y_test
```

```python
print(len(X_train))
print(len(X_test))
print(len(Y_train))
print(len(Y_test))
```

```python
model = SVC()
model.fit(X_train,Y_train)
model.score(X_test,Y_test)
```

```python
model = SVC(C = 10)
model.fit(X_train,Y_train)
model.score(X_test,Y_test)
```

```python
model = SVC(gamma = 10)
model.fit(X_train,Y_train)
model.score(X_test,Y_test)
```
