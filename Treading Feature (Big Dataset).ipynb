# Plotting Data

```python
plt.figure(figsize=(9,8))
plt.plot(df.index,df['last_price'])
plt.title('Stock Price')
plt.xlabel('Time')
plt.ylabel('Last_Price')
plt.show()
```

```python
df['rolling_mean'] = df['last_price'].rolling(window=5).mean()
df['rolling_std'] = df['last_price'].rolling(window=5).std()
df['rolling_sum'] = df['last_price'].rolling(window = 5).sum()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,mean_absolute_error,mean_squared_error
```

```python
df = df.dropna()
```

```python
x = df.drop(['last_price','date_format'],axis = 'columns')
y = df.last_price
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size = 0.3)
```

```python
print(len(X_train))
print(len(X_test))
print(len(Y_train))
print(len(Y_test))
```

```python
model = LinearRegression()
model.fit(X_train,Y_train)
model.score(X_test,Y_test)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
mean_absolute_error(Y_test,y_pred)
```

```python
mean_squared_error(Y_test,y_pred)
```
