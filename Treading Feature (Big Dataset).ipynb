# Converting date time in a format

```python
from datetime import datetime
df['created_at'] = pd.to_datetime(df['created_at'])
df['created_at']
```

```python
df['created_at'] = df['created_at'].dt.floor('S')
```

```python
#df.set_index('created_at',inplace = True)
```

```python
df['date_format'] = df['created_at'].dt.strftime('%Y-%m-%d')
```

```python
df['hour'] = df['created_at'].dt.hour
df['minut'] = df['created_at'].dt.minute
df['second'] = df['created_at'].dt.second
```

```python
# Create a new column 'Price_Status' and initialize with 0
df['Price_Status'] = 0
# Condition for price decrease
decrease_condition = df['last_price'] < df['last_price'].shift(1)
df.loc[decrease_condition, 'Price_Status'] = 1
# Condition for price increase
increase_condition = df['last_price'] > df['last_price'].shift(1)
df.loc[increase_condition, 'Price_Status'] = 2
```

```python
df['created_at'] = df['created_at'].dt.strftime('%H-%M-%S')
```

```python
def Model_LSTM():
    model = keras.Sequential([keras.layers.Input(shape=(164,1)),
                              keras.layers.LSTM(64,activation = 'tanh'),
                              keras.layers.Dense(1,activation = 'sigmoid')],
                             name = "LSTM_Model")
    model.compile(optimizer = 'adam', loss='categorical_crossentropy',metrics = ['accuracy'])
    return model                 
```
